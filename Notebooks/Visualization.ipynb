# Visualization

Matplotlib is a plotting library for Python. It includes a set of plotting commands to similar to Matlab called pyplot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's first generate some sample data

```python
random_normal = np.random.randn(100)
x = np.arange(100)
y = x**2+np.random.randn(100)*1000
```

## Let's look at some plots

### Scatter Plot
Good for looking at the relationship between numeric values

```python
plt.scatter(x,y)
```

We can all a title and lables

```python
plt.scatter(x,y)
plt.title('Relationship between x and y')
plt.xlabel('X')
plt.ylabel('Y')
```

### Histogram
A histogram is good for looking at the distribution of numberical data

```python
plt.hist(random_normal)
```

We can change the bin number and add labels

```python
plt.hist(random_normal, bins=25)
plt.title('Samples from Standard Normal')
```

### Boxplot
Boxplots are useful for looking at the distribution and key statistics of numerical data
* Median
* Outliers 1.5*IQR
* Min and Max Values
* First and Third Quartiles

```python
plt.boxplot(random_normal)
```

```python
from scipy.stats import linregress

x1 = [10.0, 8.0,  13.0,  9.0,  11.0, 14.0, 6.0,  4.0,  12.0,  7.0,  5.0]
y1 = [8.04, 6.95, 7.58,  8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]

x2 = [10.0, 8.0,  13.0,  9.0,  11.0, 14.0, 6.0,  4.0,  12.0,  7.0,  5.0]
y2 = [9.14, 8.14, 8.74,  8.77, 9.26, 8.10, 6.13, 3.10, 9.13,  7.26, 4.74]

x3 = [10.0, 8.0,  13.0,  9.0,  11.0, 14.0, 6.0,  4.0,  12.0,  7.0,  5.0]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15,  6.42, 5.73]

x4 = [8.0,  8.0,  8.0,   8.0,  8.0,  8.0,  8.0,  19.0,  8.0,  8.0,  8.0]
y4 = [6.58, 5.76, 7.71,  8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

xmax = 20
ymax = 14

print("slope, Intercept, r_value, p_value, std_error")
slope, intercept, r_value, p_value, std_err = linregress(x1, y1)
print('x1, y1')
print("{0:.3f} {1:.2f} {2:.3f} {3:.3f} {4:.3f}".format(slope, intercept, r_value, p_value, std_err))

slope, intercept, r_value, p_value, std_err = linregress(x2, y2)
print('x2, y2')
print("{0:.3f} {1:.2f} {2:.3f} {3:.3f} {4:.3f}".format(slope, intercept, r_value, p_value, std_err))

slope, intercept, r_value, p_value, std_err = linregress(x3, y3)
print('x3, y3')
print("{0:.3f} {1:.2f} {2:.3f} {3:.3f} {4:.3f}".format(slope, intercept, r_value, p_value, std_err))

slope, intercept, r_value, p_value, std_err = linregress(x4, y4)
print('x4, y4')
print("{0:.3f} {1:.2f} {2:.3f} {3:.3f} {4:.3f}".format(slope, intercept, r_value, p_value, std_err))
```

```python
ax1 = plt.subplot(2, 2, 1)
plt.scatter(x1, y1)
slope, intercept, r_value, p_value, std_err = linregress(x1, y1)
plt.plot([0, xmax], [intercept, slope * xmax + intercept])

plt.subplot(2, 2, 2, sharex=ax1, sharey=ax1)
plt.scatter(x2, y2)
slope, intercept, r_value, p_value, std_err = linregress(x2, y2)
plt.plot([0, xmax], [intercept, slope * xmax + intercept])

plt.subplot(2, 2, 3, sharex=ax1, sharey=ax1)
plt.scatter(x3, y3)
slope, intercept, r_value, p_value, std_err = linregress(x3, y3)
plt.plot([0, xmax], [intercept, slope * xmax + intercept])

plt.subplot(2, 2, 4, sharex=ax1, sharey=ax1)
plt.scatter(x4, y4)
slope, intercept, r_value, p_value, std_err = linregress(x4, y4)
plt.plot([0, xmax], [intercept, slope * xmax + intercept])

plt.axis([0, xmax, 0, ymax])
```
