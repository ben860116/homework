# Gaussian Distribution

```python
def plot_normal(mu, sigma, nsample, bins):
    import pylab as P
    P.figure()
    n, bins, patches = P.hist(np.random.normal(mu, sigma, 2000), bins, normed=1, histtype='bar', rwidth=0.8)
    P.show()
```

```python
mu = 0
sigma = 1
bins = np.linspace(-10,10,100)
plot_normal(mu, sigma, 300, bins)
```

```python
mu = 10
sigma = 1
bins = np.linspace(-10,20,100)
plot_normal(mu, sigma, 300, bins)
```

```python
mu = 0
sigma = 5
bins = np.linspace(-10,10,100)
plot_normal(mu, sigma, 300, bins)
```

```python
def plot_poisson(lam, nsample, bins):
    import pylab as P
    P.figure()
    n, bins, patches = P.hist(np.random.poisson(lam, nsample), bins, normed=1, histtype='bar', rwidth=0.8)
    P.show()
```

```python
bins = np.linspace(0,10,10)
plot_poisson(3, 6000, bins)
```

```python
bins = np.linspace(0,10,10)
plot_poisson(2, 6000, bins)
```

```python
bins = np.linspace(0,10,10)
plot_poisson(5, 6000, bins)
```

```python
bins = np.linspace(0,10,10)
plot_poisson(0, 6000, bins)
```

```python
bins = np.linspace(0,10,10)
plot_poisson(1, 6000, bins)
```

# Central Limit Theorem

```python
def plot_means(n, lam, nsample):
    avgs = []
    for i in range(n):
        avgs.append(np.mean(np.random.poisson(lam, nsample)))
    import pylab as P
    P.figure()
    bins = np.linspace(1,4,100)
    n, bins, patches = P.hist(avgs, bins, normed=1, histtype='bar', rwidth=0.8)
    P.show()
```

```python
plot_means(1,2,100)
```

```python
plot_means(10,2,100)
```

```python
plot_means(100,2,100)
```

```python
plot_means(100,2,100)
```

```python
plot_means(1000,2,100)
```

```python
plot_means(10000,2,100)
```

# Log-Normal Distribution

```python
def plot_log_normal(mu, sigma, nsample, bins):
    import pylab as P
    P.figure()
    n, bins, patches = P.hist(np.random.lognormal(mu, sigma, 2000), bins, normed=1, histtype='bar', rwidth=0.8)
    P.show()
```

```python
mu = 0
sigma = 1
bins = np.linspace(-10,10,100)
plot_log_normal(mu, sigma, 300, bins)
```

```python
mu = 5
sigma = 3
bins = np.linspace(-10,10,100)
plot_log_normal(mu, sigma, 300, bins)
```
