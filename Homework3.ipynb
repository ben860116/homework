 What is the Naive Bayes assumption? Why do we need to make this assumption?
 Naive Bayes classifier assume that the effect of the value of a predictor (x) on a given class (c) is independent of the values of other predictors. This assumption is called class conditional independence.
    We discussed the following 3 different Naive Bayes implementations in SKLearn. When would you use each?

        Bernoulli?  Features are independent booleans describig inputs.
        Multinomial?  With a multinomial event model, samples (feature vectors) represent the frequencies with which certain events have been generated by a multinomial (p_1, \dots, p_n) where p_i is the probability that event i occurs (or K such multinomials in the multiclass case).
